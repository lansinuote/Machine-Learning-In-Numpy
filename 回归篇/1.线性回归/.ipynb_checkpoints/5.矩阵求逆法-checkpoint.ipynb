```python
import numpy as np


#加载数据
def load_data():
    with open('多变量线性数据.txt') as fr:
        lines = fr.readlines()

    #在x当中多加一列,全部为0,用来计算b
    x = np.ones((len(lines), 3), dtype=float)
    y = np.empty((len(lines), 1), dtype=float)

    for i in range(len(lines)):
        line = lines[i].strip().split(',')
        x[i, :2] = line[:2]
        y[i] = line[2]

    #这里给他填充点随机数,不然会导致x*x是奇异矩阵,无法求逆
    x[:, 0] = np.random.rand(len(x))

    #转矩阵
    return np.mat(x), np.mat(y)


x, y = load_data()
x[:5], y[:5]
```

```python
#常量
N, M = x.shape
#现在的w当中包括了w和b
w = np.mat([[0], [1], [0]], dtype=float)
w
```

```python
#预测函数
def predict(x):
    return (x * w)[0, 0]


predict(x[0])
```

```python
#求loss,就是简单的做差求平方和
def get_loss():
    loss = (y - x * w).T * (y - x * w)
    return loss[0, 0]


get_loss()
```

```python
#这个公式是从书上抄的,没推算过
def get_w():
    xx = x.T * x
    
    #求内积的行列式值,如果是奇异矩阵就不能求逆,就不能用这种方法了
    if np.linalg.det(xx) == 0:
        raise Exception('奇异矩阵,无法求逆')

    w = xx.I * x.T * y
    return w


w = get_w()
w
```

```python
for xi, yi in zip(x, y):
    print(predict(xi), yi)
```
